```python
import csv
import json

def csv_to_json(csv_file_path, json_file_path):
    # Open the CSV file
    with open(csv_file_path, 'r', encoding='utf-8-sig') as csv_file:  # Remove BOM
        # Read the CSV file using the DictReader
        csv_reader = csv.DictReader(csv_file)

        # Initialize an empty list to store data
        data = []

        # Iterate over each row in the CSV
        for row in csv_reader:
            # Create a dictionary to store each row's data
            row_data = {key: value for key, value in row.items() if key}  # Filter out empty keys
            data.append(row_data)

    # Remove any empty dictionaries from the list
    data = [row for row in data if row]

    # Write the data to a JSON file
    with open(json_file_path, 'w') as json_file:
        # Indent the JSON for readability
        json.dump(data, json_file, indent=4)

# Example usage:
csv_to_json('/content/Interview_Feedbacks - Feuille 1.csv', '/content/drive/MyDrive/Feedback Dataset/interview_feedback.json')
```
